Бинарный поиск

```python
def bianry_search(n:int) -> int:
    low_threshold = 0
    high_threshold = 1

    while high_threshold < n:
        high_threshold *= 10

    mid = (high_threshold + low_threshold) // 2

    while mid != n:
        mid = (high_threshold + low_threshold) // 2
        
        if n > mid:
            low_threshold = mid

        else:
            high_threshold = mid

    return mid
```

Бинарное дерево

```python
class BSTNode:
    def __init__(self, val:int=None) -> None:
        self.left = None
        self.right = None
        self.val = val

    def insert(self, val:int) -> None:
        if not self.val:
            self.val = val
            return

        if self.val == val:
            return

        if val < self.val:
            if self.left:
                self.left.insert(val)
                return
            
            self.left = BSTNode(val)
            return

        if self.right:
            self.right.insert(val)
            return
        
        self.right = BSTNode(val)

    def search(self, val:int) -> int:
        if val == self.val:
            return val

        if val < self.val:
            if self.left == None:
                return False
            
            return self.left.search(val)

        if self.right == None:
            return False
        
        return self.right.search(val)

    def inorder(self, vals:int) -> List[int]:
        if self.left is not None:
            self.left.inorder(vals)

        if self.val is not None:
            vals.append(self.val)

        if self.right is not None:
            self.right.inorder(vals)

        return vals
    
    def delete(self, val:int) -> None:
        if self == None:
            return self
        
        if val < self.val:
            if self.left:
                self.left = self.left.delete(val)

            return self
        
        if val > self.val:
            if self.right:
                self.right = self.right.delete(val)

            return self
        
        if self.right == None:
            return self.left
        
        if self.left == None:
            return self.right
        
        min_larger_node = self.right
        
        while min_larger_node.left:
            min_larger_node = min_larger_node.left

        self.val = min_larger_node.val
        self.right = self.right.delete(min_larger_node.val)

        return self
```

Поиск по Фиббоначчи

```python
def fibbo_search(target:int) -> int:
    if target < 0:
        return False
    
    elif target == 0:
        return 0
    
    fib1 = 0
    fib2 = 1
    sum_fib = 1
    
    while sum_fib <= target:
        sum_fib = fib1 + fib2
        fib2, fib1 = fib1, sum_fib

    while fib2 != target:
        fib2 += 1

    return fib2
```

Интерполяционный поиск

```python
def interpolation_search(arr:List[int], target:int) -> int: 
    if not arr:
        return -1
 
    left, right = 0, len(arr) - 1
 
    while arr[right] != arr[left] and arr[left] <= target <= arr[right]:
        mid = left + (target - arr[left]) * (right - left) // (arr[right] - arr[left])
 
        if target == arr[mid]:
            return mid

        elif target < arr[mid]:
            right = mid - 1

        else:
            left = mid + 1
 
    if target == arr[left]:
        return left

    return -1
```

```python
hash_table = [None] * 5

def hash_func(key:int, leng:int) -> int:
    return key % leng

def insert(key:int, value:int) -> int:
    global hash_table
    length = len(hash_table)
    index = hash_func(key, length)

    if  None not in hash_table:
        new_hash_table = [None] * len(hash_table) * 2
        
        for i in range (len(hash_table)):
            new_hash_table[i] = hash_table[i]

        hash_table = copy(new_hash_table)

    while hash_table[index] is not None:
        index = (index + 1) % len(hash_table)

    hash_table[index] = value

    
insert(1, 'A')
insert(5, 'B')
insert(12, 'C')
insert(9, 'D')
print(hash_table)

#вычисление нового индекса
insert(4, 'E')
print(hash_table)

#расширение
insert(3, 'G')
print(hash_table)

insert(9, "H")
print(hash_table)
```

```python
numbers = [4, 8, 15, 16, 23, 42]
new_numbers = []
np.random.seed(2104)

for number in numbers:
    random_number = np.random.randint(1, len(numbers) + 1)
    hashed_number = (number * random_number) % (len(numbers) + 2)
    new_numbers.append(hashed_number)

print("Исходный список чисел:", numbers)
print("Перехешированный список чисел:", new_numbers)

def search_rand(n:int, tar:int) -> bool:
    np.random.seed(2104)

    for _ in range(n + 1):
        random_number = np.random.randint(1, len(numbers) + 1)

    check_number = (tar * random_number) % (len(numbers) + 2)
    
    return check_number == new_numbers[n]

def alt_search_rand(tar:int) -> bool:
    global new_numbers
    np.random.seed(2104)
    
    for _ in range(len(new_numbers)):
        random_number = np.random.randint(1, len(numbers) + 1)
        check_number = (tar * random_number) % (len(numbers) + 2)
        
        if check_number in new_numbers:
            return True
    
    return False

target = np.random.choice(numbers)
ind = numbers.index(target)
print(f"{target} находится в перехэшированном списке: {search_rand(ind, target)}")
print(f"{target} находится в перехэшированном списке: {alt_search_rand(target)}")
```

```python
class Node:
    def __init__(self, key:int, value:int) -> None:
        self.key = key
        self.value = value
        self.next = None

class HashTable:
    def __init__(self) -> None:
        self.capacity = 5
        self.size = 0 
        self.buckets = [None] * self.capacity 
  
    def add(self, key:int, value:int) -> None:
        index = hash(key) % self.capacity
        node = self.buckets[index]

        while node: 
            if node.key == key:
                node.value = value

                return
            
            node = node.next
        
        new_node = Node(key, value)
        new_node.next = self.buckets[index]
        self.buckets[index] = new_node
        self.size += 1
        
        if self.size > 0.7 * self.capacity:
            self.expand()
  
    
    def get(self, key:int) -> None:
        index = hash(key) % self.capacity 
        node = self.buckets[index] 

        while node: 
            if node.key == key:
                return node.value
            
            node = node.next
        
        return None
  
    
    def delete(self, key:int) -> None:
        index = hash(key) % self.capacity 
        node = self.buckets[index]
        prev = None

        while node: 
            if node.key == key:
                if prev:
                    prev.next = node.next

                else:
                    self.buckets[index] = node.next

                self.size -= 1
                return
            
            prev = node
            node = node.next
  
    def expand(self) -> None:
        new_capacity = self.capacity * 2 
        new_buckets = [None] * new_capacity 
        
        for i in range(self.capacity):
            node = self.buckets[i]

            while node:
                index = hash(node.key) % new_capacity

                if new_buckets[index]:
                    new_node = new_buckets[index]

                    while new_node.next:
                        new_node = new_node.next

                    new_node.next = Node(node.key, node.value)

                else:
                    new_buckets[index] = Node(node.key, node.value)

                node = node.next
                
        self.capacity = new_capacity
        self.buckets = new_buckets

ht = HashTable()
ht.add('key1', 'value1')
ht.add('key2', 'value2')
ht.add('key3', 'value3')
ht.add('key4', 'value4')
ht.add('key1', 'value11')
print(ht.get('key1'))
ht.delete('key3')
print(ht.get('key3'))
```

### Задание 3
Ферзи

```python
num_queens = 8 
current_solution = [0 for _ in range(num_queens)]
solutions = [] 

def is_safe(temp_row:int, temp_col:int) -> bool:
    if temp_row == 0:
        return True

    for row in range(0, temp_row):
        if temp_col == current_solution[row]:
            return False

        if abs(temp_row - row) == abs(temp_col - current_solution[row]):
            return False

    return True

def place_queen(row:int) -> None:
    global current_solution, solutions, num_queens

    for col in range(num_queens):
        if not is_safe(row, col):
            continue
        
        else:
            current_solution[row] = col
            if row == (num_queens - 1):
                solutions = current_solution

            else:
                place_queen(row + 1)
                if len(solutions) >= 1:
                    break

place_queen(0)
print(solutions)
```
